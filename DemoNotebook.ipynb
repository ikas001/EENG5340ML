```python
my_array=np.array([2,3,7.5]) #Declare our x vector
my_matrix=np.array([[1,2,-4],[3.5,.5,-.7],[-3.2,2,2.5]]) #Declare our matrix
x=np.dot(my_matrix,my_array) #create b from the matrix-vector product
x_hat=np.linalg.solve(my_matrix,my_array) #see if we can recover
x-x_hat
```

So, I have deliberately(?) left an error there (because we do not have a vector of zeros).  Can you fix it?  HINT: should be very easy. Let's move on to defining some simple functions.  This is simple numerical differentiation that will take on two values, x and h, and calculate the derivative of a function at that point.  This simple first order formula is well known (but not particularly accurate) $$\frac{df}{dt}\approx\frac{f(t+h)-f(t)}{t}$$.  For small enough h this should work great.  We will see.  

```python
def f(x): #This part is easy
    return(np.sin(x)+x) #Pay attention to indentation and the fact that the sine function is part of numpy

def f_prime(x,h): #define our naive derivative
    return((f(x+h)-f(x))/h)
```

```python
print f_prime(1,.1),f_prime(1,.01),f_prime(1,.001)
```

So the approximation for h=0.1 is pretty crummy as you would expect.  h values of 0.01 and 0.001 are much better.  The actual solution is, of course:

```python
1+np.cos(1) #don't forget the numpy namespace
```

How small should h be?  Well this is a small toy problem, but I would not feel comfortable for values greater than 1e-6

```python
f_prime(1,1e-6)
```

```python
x=np.linspace(-3,3,50) #Just create a linear space in 1D
y=np.linspace(-3,3,50)  #same
print x.shape #Let's verify the number of points and values
print np.amin(x),np.amax(x)
```

```python
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
xx,yy=np.meshgrid(x,y) #Define a mesh
z=-(xx**2+yy**2)
ax.plot_wireframe(xx,yy,z)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```
